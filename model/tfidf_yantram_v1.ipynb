```python
df = df['Translation'] # we are using only data from translation column of the csv(excel)
corpus=df.to_list()
```

```python
# This part of code actually finds the tfidf score for each word using 'TfidfVectorizer' function

gita_txt = corpus
gita_txt = [i.translate(str.maketrans('', '', string.punctuation)) for i in corpus] # remove unnecessary punctuations
tf = TfidfVectorizer(smooth_idf=False, sublinear_tf=False, norm=None, analyzer='word',use_idf=True)
txt_fitted = tf.fit(gita_txt)
txt_transformed = txt_fitted.transform(gita_txt) # this is a sparse matrix containing tfjidf scores for each document
print ("The text: ", gita_txt)
```

```python
vocab = tf.vocabulary_

# this function is used to find score for each query with respect to each document

def find_score(query):
    doc_score_list = []
    doc_index_list = []
    for i in range(0,len(gita_txt)):
        document_1_words = query.split()
        document_2_words = gita_txt[i].split()
        
        common = set(document_1_words).intersection( set(document_2_words) )# stores common words between query and current document
        common = [ i for i in common if i in vocab.keys()]  # stores the common word if they are present in our tfidf vocabulary, some common words might have been removed from vocabulary by TfidfVectorizer
        
        myKeys = [vocab[words.lower()] for words in common]  # get key(intger values) corresponding to each common word
        doc_score = df.iloc[myKeys,i].sum()  # Find the score of current document with respect to the query
#       print(myKeys)

        if common:
            #print(i,"common ",common, 'score= ',doc_score) # uncomment this to find common words between query and current document, useful for debuggin and improving solution
            doc_score_list.append(doc_score)
            doc_index_list.append(i)

    # sort the document scores
    sorted_indices = [x for _,x in sorted(zip(doc_score_list,doc_index_list), reverse=True)]
    
    # return top K (K=3) results, i.e, return top 3 doc score results as asked in submission output format
    K = 3
    sorted_indices_top_K = sorted_indices[:K]
       
    return sorted_indices_top_K,sorted(doc_score_list,reverse=True)

```
