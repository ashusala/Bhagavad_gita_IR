```python
from selenium import webdriver
from bs4 import BeautifulSoup
import unidecode
```

```python
# setting up our chrome webdriver
driver = webdriver.Chrome("/usr/local/bin/chromedriver")

verse_number = []
translations = []

# launching test chrome on chapter1-18 
# need a loop here to change pages automatically
for num in range(1,19):
    fixed_part = "https://vedabase.io/en/library/bg/"
    dynamic = fixed_part + str(num) + '/'
    driver.get(dynamic)
    
    content = driver.page_source
    soup = BeautifulSoup(content)
    
    for verse in soup.find_all('div', attrs={'class':'rich-text'}):
        if verse.find('dd') != None:
            translations.append(unidecode.unidecode(verse.find('dd').text))
            verse_number.append(verse.find('dt').text.split()[-1])
            verse_number[-1] = str(num) + '.' + verse_number[-1]
```

```python
import pandas as pd
df = pd.DataFrame(zip(verse_number,translations), columns=['#verse','Translation'])
df.to_csv('data_translations.csv')
```

```python
# FUNCTION to get the purports out of a page in a list called verses.
def find_purports(purports, content):
    soup = BeautifulSoup(content)
    soup = soup.find('div',attrs={'class':'wrapper-puport'})
    if soup is None: return purports
    for pur in soup.find_all('div', attrs={'class':'rich-text'}):
        if pur.find('p') != None:
            purports.append(pur.find('p').text)
```
